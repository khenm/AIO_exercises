### **Data Analysis with IMDB Movie Data**
IMDB Movie Dataset is a dataset evaluating the movie by multiple criteria including directors, actors and genres. The dataset aims to predict the rating of the movie. The following section will handle the data analysis using pandas.

```python
#@title Import Libraries and Load Dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset_path = 'IMDB-Movie-Data.csv'
df = pd.read_csv(dataset_path)
df.head()
```

```python
# Read data with specified explicit index.
data_indexed = pd.read_csv(dataset_path, index_col='Title')
data_indexed.head()
```

```python
#@title The basic information about this data
df.info()
```

```python
df.describe()
```

```python
#@title Data Selection - Indexing and Slicing data
# Extract the data as series
genre = df['Genre']
genre.head()
```

```python
# Extract data as database
df[['Genre']].head()
```

```python
# We can choose multiple columns to make a new dataframe
some_cols = df[['Title', 'Genre', 'Actors', 'Director', 'Rating']]
some_cols.head()
```

```python
df.iloc[10:15][['Title', 'Rating', 'Revenue (Millions)']]
```

```python
# Data Selection - Based on Conditional Filtering
df[((df['Year'] >= 2010) & (df['Year'] <= 2015))
   & (df['Rating'] < 6.0)
   & (df['Revenue (Millions)'] > df['Revenue (Millions)'].quantile(0.95))]
```

```python
#@title Groupby Operations
df.groupby('Director')[['Rating']].mean().head()
```

```python
#@title Sorting Operations
df.groupby('Director')[['Rating']].mean().sort_values(['Rating'], ascending=False).head()
```

```python
#@title Missing Values
df.isnull().sum()
```

```python
# Deal with missing values - Deleting
df.drop('Metascore', axis=1).head()
```

```python
# Dealing with missing values - Filling
revenue_mean = data_indexed['Revenue (Millions)'].mean()
print('The mean revenue is: ', revenue_mean)

data_indexed['Revenue (Millions)'].fillna(revenue_mean, inplace=True)
```

```python
#@title Apply Functions
""" Apply function is used when we want to execute a function on each row/ each column of a dataframe. This offer the same in Excel or Google Sheet functions.
"""
def rating_group(rating):
  if rating >= 7.5:
    return 'Good'
  elif rating >= 6.0:
    return 'Average'
  else:
    return 'Bad'

df['Rating_category'] = df['Rating'].apply(rating_group)
df[['Title', 'Director', 'Rating', 'Rating_category']].head()
```
