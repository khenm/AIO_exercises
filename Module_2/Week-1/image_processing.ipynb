# Image Processing
Converting Image from RGB color channels into grayscale.

The method can be done by computing the RGB values to get only 1 value from
0 to 255. There are 3 ways to do it: \
- Lightness: compute the average value of max and min RGB
- Average: compute the average value of all RGB
- Luminosity: Multiply by the coefficient: 0.21 * R + 0.72 * G + 0.07 * B



```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import os
```

```python
current_dir = os.getcwd()
folder_path = os.path.join(current_dir, 'sample')
file_path = folder_path + '\dog.jpeg'
```

1. Lightness Method

```python
def color2grayscale(vector):
    max = np.max(vector).astype(float)
    min = np.min(vector).astype(float)
    result = (max + min) / 2
    return result

def lightness(img_path):
    """converting image from RGB color channels into grayscale by
    computing the average value of max and min RGB.
    Args:
        img_path: input image
    Returns
        gray: grayscale image
    """
    img = mpimg.imread(img_path)
    gray = np.apply_along_axis(color2grayscale, 2, img)
    return gray

gray = lightness(file_path)
plt.imshow(gray.astype(np.uint8), cmap='gray')
```

2. Average Method

```python
def color2grayscale(vector):
    result = vector.mean()
    return result

def average(img_path):
    """Converting image from RGB color channels into grayscale by
    computing the average value of all RGB.
    Args:
        img_path: input image
    Returns
        gray: grayscale image
    """
    img = mpimg.imread(img_path, 1)
    gray = np.apply_along_axis(color2grayscale, 2, img)
    return gray

gray = average(file_path)
plt.imshow(gray.astype(np.uint8), cmap='gray')
```

3. Luminosity Method

```python
def luminosity(img_path):
    """Converting image from RGB color channels into grayscale by
    multiplying by the coefficient: 0.21 * R + 0.72 * G + 0.07 * B.
    Args:
        img_path: input image
    Returns
        gray: grayscale image
    """
    img = mpimg.imread(img_path)
    R = img[:, :, 0]
    G = img[:, :, 1]
    B = img[:, :, 2]
    gray = 0.21 * R + 0.72 * G + 0.07 * B
    return gray

gray = luminosity(file_path)
plt.imshow(gray.astype(np.uint8), cmap='gray')
```
