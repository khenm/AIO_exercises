### Tabular Data Analysis

```python
import numpy as np
import pandas as pd

data = pd.read_csv('advertising.csv')

def correlation(X, Y): 
    N = len(X)
    numerator = 0
    denominator = 0
    
    # numerator
    sum_X = np.sum(X)
    sum_Y = np.sum(Y)
    sum_X_prod_Y = np.dot(X, Y)
    numerator = N * sum_X_prod_Y - (sum_X * sum_Y)
    
    # denominator
    sum_X_squared = np.dot(X, X)
    sum_Y_squared = np.dot(Y, Y)
    denominator_X = np.sqrt(N * sum_X_squared - (sum_X ** 2))
    denominator_Y = np.sqrt(N * sum_Y_squared - (sum_Y ** 2))
    denominator = denominator_X * denominator_Y
    return np.round(numerator / denominator, 2)

X = data['TV']
Y = data['Radio']
corr_xy = correlation(X, Y)
print(f'Correlation between TV and Sales: {round(corr_xy, 2)}')
```

```python
features = ['TV', 'Radio', 'Newspaper']

for feature_1 in features:
    for feature_2 in features:
        correlation_value = correlation(data[feature_1], data[feature_2])
        print(f'Correlation between {feature_1} and {feature_2}: {round(correlation_value, 2)}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
data_corr = data.corr()
sns.heatmap(data_corr, annot=True,  fmt='.3f', linewidth=.5)
plt.show()
```
